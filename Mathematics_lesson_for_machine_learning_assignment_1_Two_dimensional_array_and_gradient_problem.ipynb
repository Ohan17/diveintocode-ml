```python
# [Problem 1] Linear function
import numpy as np
nd_x = np.arange(-50, 51)
nd_y = np.array(list(map(lambda x: 0.5*x+1, nd_x)))
print(nd_x)
print(nd_y)
```

```python
# [Problem 2] Array combination
nd_z = np.array(list(map(lambda x,y: [x,y], nd_x,nd_y)))
print(nd_z.shape)
```

```python
# [Problem 3] Find the gradient
from sympy import *
x = symbols('x')
y = symbols('y')
d_x = np.array(nd_x[1:]-nd_x[:-1])
d_y = np.array(nd_y[1:]-nd_y[:-1])
grad = np.array(list(map(lambda a,b: limit(b/x,x,a), d_x, d_y)))
print(grad)
```

```python
# [Problem 4] Draw a graph
import matplotlib.pyplot as plt
plt.xlabel('x')
plt.ylabel('y')
plt.plot(nd_x,nd_y)
```

```python
# [Problem 5] Python functionalization
# [Problem 6] Find the minimum value
def compute_gradient(function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.
    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    nd_x = np.arange(*x_range)
    nd_y = function(nd_x)
    x = symbols('x')
    y = symbols('y')
    d_x = np.array(nd_x[1:]-nd_x[:-1])
    d_y = np.array(nd_y[1:]-nd_y[:-1])
    array_xy = np.array(list(map(lambda x,y: [x,y], nd_x,nd_y)))
    gradient = np.array(list(map(lambda a,b: limit(b/x,x,a), d_x, d_y)))
    return array_xy, gradient

def function1(array_x):
    return np.array(list(map(lambda x: x**2, array_x)))
print("Function 1")
array_xy, gradient = compute_gradient(function1)
plt.title('y = x^2')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy.T[0],array_xy.T[1])
plt.show()
min_y = np.min(array_xy.T[1])
min_idx_y = np.argmin(array_xy.T[1])
print("Min y = {}".format(min_y))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:min_idx_y].T[0],array_xy[:min_idx_y].T[1])
plt.show()
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[min_idx_y:].T[0],array_xy[min_idx_y:].T[1])
plt.show()

def function2(array_x):
    return np.array(list(map(lambda x: 2*x**2+2**x, array_x)))
print("Function 2")
array_xy, gradient = compute_gradient(function2)
plt.title('y = 2x^2 + 2^x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy.T[0],array_xy.T[1])
plt.show()
min_y = np.min(array_xy.T[1])
min_idx_y = np.argmin(array_xy.T[1])
print("Min y = {}".format(min_y))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:min_idx_y].T[0],array_xy[:min_idx_y].T[1])
plt.show()
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[min_idx_y:].T[0],array_xy[min_idx_y:].T[1])
plt.show()

def function3(array_x):
    return np.array(list(map(lambda x: np.sin(x**0.5), array_x)))
print("Function 3")
array_xy, gradient = compute_gradient(function3, (0, 50.1, 0.1))
plt.title('y = sin(x^0.5)')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy.T[0],array_xy.T[1])
plt.show()
min_y = np.min(array_xy.T[1])
min_idx_y = np.argmin(array_xy.T[1])
print("Min y = {}".format(min_y))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[:min_idx_y].T[0],array_xy[:min_idx_y].T[1])
plt.show()
plt.xlabel('x')
plt.ylabel('y')
plt.plot(array_xy[min_idx_y:].T[0],array_xy[min_idx_y:].T[1])
plt.show()
```
