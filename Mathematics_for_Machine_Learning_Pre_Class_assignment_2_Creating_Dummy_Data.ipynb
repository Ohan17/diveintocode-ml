```
# [Problem 1] Creating random numbers
import numpy as np
import matplotlib.pyplot as plt
mean = [-3, 0]
cov = [[1., 0.8], [0.8, 1.]]
np.random.seed(0)
data1 = np.random.multivariate_normal(mean, cov, 500)
print("Array shape: {}.".format(data1.shape))
plt.plot(data1.T[0], data1.T[1], 'x')
plt.axis('equal')
plt.show()
```

```
# [Problem 2] Visualization by scatter plot
plt.scatter(data1.T[0], data1.T[1])
```

```
# [Problem 3] Visualization by histogram
plt.title('histogram of x')
plt.xlabel('frequency')
plt.ylabel('value')
plt.hist(data1.T[0], label='x')
plt.xlim(-6, 3)
plt.show()

plt.title('historgram of y')
plt.xlabel('frequency')
plt.ylabel('value')
plt.hist(data1.T[1], label='y')
plt.xlim(-6, 3)
plt.show()
```

```
# [Problem 4] Addition of data
mean = [0, -3]
cov = [[1, 0.8], [0.8, 1]]
np.random.seed(1)
data2 = np.random.multivariate_normal(mean, cov, 500)

plt.scatter(data1.T[0], data1.T[1], label='0')
plt.scatter(data2.T[0], data2.T[1], label='1')
plt.legend()
```

```
# [Problem 5] Data combination
data = np.concatenate((data1, data2))
print("Concatenated array shape: {}.".format(data.shape))
```

```
# [Problem 6] Labeling
a = np.concatenate((data1, np.zeros((len(data1), 1))), axis=1)
b = np.concatenate((data2, np.ones((len(data1), 1))), axis=1)
data = np.concatenate((a,b))
print("Concatenated array shape: {}.".format(data.shape))
print(data)
```
