```
import time
# [Problem 1]
"""
Code to calculate the thickness when the paper is folded 43 times
"""
THICKNESS = 0.00008

start = time.time()
folded_thickness = THICKNESS*2**43
elapsed_time = time.time() - start

print("Thickness: {} meters".format(folded_thickness))
print("time : {}[s]".format(elapsed_time))

# [Problem 2]
# Convert meters to 10,000 kilometers and display with two decimal places
print("Thickness: {: .2f} 10,000 kilometers".format(folded_thickness/10000000))
```

```
# [Problem 3]
THICKNESS = 0.00008
folded_thickness = THICKNESS

start = time.time()
for n in range(43): 
  folded_thickness *= 2
elapsed_time = time.time() - start

print("Thickness: {} meters".format(folded_thickness))
print("time : {}[s]".format(elapsed_time))
```

```
# [Problem 4]
"""
After a few runs, the for statement approach has better calculation speed compare to using exponential directly.
Eventhough using exponent may be easier to read and save coding time, larger iteration size may have to use the
problem 2 approach to save calculating time.
"""
```

```
# [Problem 5]

THICKNESS = 0.00008
folded_thickness = THICKNESS
record = [THICKNESS]

for n in range(43): 
  folded_thickness *= 2
  record.append(folded_thickness)

print("Thickness: {: .2f} kilometers".format(folded_thickness/1000))
print("Record length: {}".format(len(record)))

# [Problem 6]
"""
Display the graph. Title and axis label name.
"""
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(record) # Enter the variable name of the list in "List name"
plt.show()

# [Problem 7]
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(record, color='black',marker='.', markerfacecolor='white', markersize=5) # Enter the variable name of the list in "List name"
plt.show()

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.tick_params(labelsize=10)
plt.plot(record, ls="-.",) # Enter the variable name of the list in "List name"
plt.show()

plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(record, lw=5.0) # Enter the variable name of the list in "List name"
plt.show()
```
